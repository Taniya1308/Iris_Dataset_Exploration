```python
#import the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# print(enumerate(iris.target_names))
target_names = dict(enumerate(iris.target_names))
```

```python
print(target_names)
```

```python
df["Species"] = df["target"].map(target_names)
```

```python
#quick Pandas inspection
print("\n shape", df.shape)
print("-"*50)
print("\n Info:", df.info())
print("-"*50)
print("\n Describe (numeric):", df.describe())
```

```python
# convert to numpy arrays
feature_cols = iris.feature_names #it contain all 3 flowers nuiuj
x=df[feature_cols]
y=df["target"]
```

```python
print(type(x))
```

```python
print(type(y))
```

```python
print(x)
```

```python
print(y)  #y is having one column that's why it is a series
```

```python
x=x.to_numpy()
y=y.to_numpy()
```

```python
# NumPy stats (column-wise): mean, std
col_means= x.mean(axis=0)  #axis = 0 → operate column-wise
# ddof = Delta Degrees Of Freedom
col_stds= x.std(axis=0, ddof=0) # population std (ddof=0)
print("\n NumPy column means:", col_means)
print("Numpy column stds:", col_stds)
```

```python
# find missing value in dataset
print("Missing value in the dataframe")
print(df.isnull().sum())
```

```python
# aggregate data
df.groupby("Species")[feature_cols].mean()
```

```python
# pair plot with seaborn
sns.set_style("whitegrid")

sns.pairplot(df[feature_cols + ["Species"]], hue="Species", diag_kind="kde")
plt.suptitle("Pairwise feature relationship")
plt.show()

# feature_cols + ["Species"]
```

```python
# correlation heatmap
corr=df[feature_cols].corr(numeric_only=True)
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation heatmap")
plt.show()
# print(corr)
```

```python
# (C) Class distribution bar chart
plt.figure(figsize=(6, 4))
sns.countplot(x="Species", data=df)
plt.title("Class Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Simple scatter with hue (example:petal length vs petal width)
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="Species"
)
plt.title("Petal length vs petal width by Species")
plt.tight_layout()
plt.show()
```
